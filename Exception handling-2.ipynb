Q1. Explain why we have to use the Exception class while creating a Custom Exception.

A1. To improve readability of your code.
To enhance reusability of features.
To provide custom messages/instructions to users for specific use cases.

```python
class Error(Exception):
    pass
 
class TransitionError(Error):
    def __init__(self, prev, nex, msg):
        self.prev = prev
        self.next = nex
        self.msg = msg
try:
    raise(TransitionError(2, 3*2, "Not Allowed"))
except TransitionError as error:
    print('Exception occurred: ', error.msg)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

A3. ArithmeticError is simply an error that occurs during numeric calculations. ArithmeticError types in Python include: OverFlowError. ZeroDivisionError.

```python
try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

A4. LookupError Exception is the Base class for errors raised when something can't be found. The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: IndexError, KeyError. An IndexError is raised when a sequence reference is out of range.

```python
try: 
    a = [1, 2, 3] 
    print (a[3]) 
except LookupError: 
    print ("Index out of bound error.")
else: 
    print ("Success")
```
